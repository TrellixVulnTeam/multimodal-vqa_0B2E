```python
import json
import numpy as np

 
# Opening JSON file
f = 'train.json'
#f = open('../data/train.json',)
 
# returns JSON object as
# a dictionary]
lines = open(f).readlines()
examples = [json.loads(line) for line in lines]

 
# Iterating through the json
# list
sents = []
for i in examples:
    sents.append(i['sentence'])
```

```python
# import necessary libraries
import tensorflow_hub as hub
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()
  
# Load pre trained ELMo model
elmo = hub.Module("https://tfhub.dev/google/elmo/3", trainable=True)
#elmo = hub.Module("https://tfhub.dev/google/nnlm-en-dim128/1", trainable=True)
  
# create an instance of ELMo
embeddings = elmo(
    sents[:100],
    signature="default",
    as_dict=True)['elmo']
init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
  
# Print word embeddings for word WATCH in given two sentences
#print('Word embeddings for word WATCH in first sentence')
#print(sess.run(embeddings[0][3]))
#print('Word embeddings for word WATCH in second sentence')
#print(sess.run(embeddings[1][5]))
```

```python
all_embs = None
```

```python
for i in range(30):#863):
  # create an instance of ELMo
  embeddings = elmo(
      sents[100*i:100*(i+1)],
      signature="default",
      as_dict=True)['elmo']
  init = tf.initialize_all_variables()

  emb = sess.run(embeddings)
  emb.shape
  sent_embs = np.mean(emb, axis = 1)
  if(all_embs is None):
    all_embs = sent_embs
  else:
    all_embs = np.vstack((all_embs, sent_embs))
  print(i)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

X = np.array(tsne_df[['x','y']])
all_embs

kmeans = KMeans(n_clusters=10, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(all_embs)
plt.scatter(X[:,0], X[:,1],c = pred_y)
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, all_embs)
for idx in closest:
  print(sents[idx])
  plt.scatter(X[idx,0],X[idx,1], c = idx, label = sents[idx])
  plt.scatter(X[idx,0],X[idx,1], c = 'red')
print(closest)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:,0], X[:,1],c = pred_y, alpha = .5)
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, all_embs)
markers = ['x','o','c','n','e']
for idx in closest:
  print(X[idx,0],X[idx,1])
  print(sents[idx])
  plt.text(X[idx,0],X[idx,1], sents[idx],backgroundcolor = 'white' )
  plt.scatter(X[idx,0],X[idx,1], c = 'red', s = 1000)

plt.legend()
print(closest)
plt.show()
```
